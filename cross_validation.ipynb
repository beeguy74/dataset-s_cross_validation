```python
import matplotlib as mp
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.linear_model import LogisticRegression

```

```python
data = pd.read_csv("data.csv")
data.head()
```

```python
features = data.drop(["id", "Unnamed: 32", "diagnosis"], axis=1)
features.head()
```

```python
kf = KFold(n_splits=5)
model = LogisticRegression(max_iter=7000)
scores = []
```

```python
i = 0
for train_index, test_index in kf.split(features, data.diagnosis):
    X_train, X_test = features.iloc[train_index,:], features.iloc[test_index,:]
    y_train, y_test = data.diagnosis[train_index], data.diagnosis[test_index]  
    # fit the model on the training data
    model.fit(X_train, y_train)

    # evaluate the model on the testing data
    score = model.score(X_test, y_test)
    scores.append(score)
    # print the accuracy score for each fold
    print("Accuracy:", score)
```

```python
# calculate the mean accuracy score across all folds
mean_score = np.mean(scores)
print("Mean accuracy:", mean_score)
```

```python
from sklearn.model_selection import cross_val_score
import warnings
warnings.filterwarnings("ignore")

# initialize a logistic regression model

# perform 5-fold cross-validation and get the accuracy scores for each fold
scores = cross_val_score(model, features,data[['diagnosis']], cv=5)

# print the accuracy scores for each fold and the mean accuracy score
print("Accuracy scores for each fold:", scores)
print("Mean accuracy score:", scores.mean())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X_train, X_test, y_train, y_test = train_test_split(features, data.diagnosis, test_size=0.2, random_state=42)
```

```python
rf_model=RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
scores = cross_val_score(rf_model, X_train, y_train, cv=5, scoring='balanced_accuracy')
```

```python
# print the accuracy scores for each fold and the mean accuracy score
print("Precision scores for each fold:", scores)
print("Mean precision score:", scores.mean())
```

```python
from sklearn.metrics import classification_report
rf_scores=[]
for train_index, test_index in kf.split(X_train, y_train):
    X_train, X_test = features.iloc[train_index,:], features.iloc[test_index,:]
    y_train, y_test = data.diagnosis[train_index], data.diagnosis[test_index]  
    # fit the model on the training data
    rf_model.fit(X_train, y_train)
    y_pred = rf_model.predict(X_test)

    # evaluate the model on the testing data
    score = rf_model.score(X_test, y_test)
    rf_scores.append(score)
    # print the accuracy score for each fold
    print("report:", classification_report(y_test, y_pred))
```
